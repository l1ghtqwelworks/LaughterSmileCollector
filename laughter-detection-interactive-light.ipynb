```python
#@title Choose settings and run Laughter Detection

threshold = 0.8 #@param {type:"slider", min:0.1, max:1.0, step:0.1}
min_length = 0.6 #@param {type:"slider", min:0.1, max:1.0, step:0.1}
save_to_audio_files = False #@param {type:"boolean"}
save_to_textgrid = True #@param {type:"boolean"}
output_dir = 'laughter_detection_output' #@param {type:"string"}
audio_path = 'temp_audio.wav'
```

```python
import laugh_segmenter_light

instances = laugh_segmenter_light.load_and_segment_laughter(audio_path, threshold, min_length)
```

```python
print(); print("found %d laughs." % (len (instances)))
print(instances)
```

```python
## write predictions

laugh_segmenter_light.save_laughter_segments(instances, audio_path, output_dir, save_to_audio_files, save_to_textgrid)
```

```python
#@title Listen to the detected laughter

import IPython
import librosa
from IPython.display import Audio
detected_laughs = librosa.util.find_files(output_dir)
for laugh in detected_laughs:
  IPython.display.display(Audio(laugh))
```
